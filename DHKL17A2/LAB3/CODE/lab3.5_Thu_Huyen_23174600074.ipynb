```python
import pandas as pd
```

```python
stocks1=pd.read_csv('D:\Python nâng cao\Thu_Huyen_23174600074_17A2\DHKL17A2\LAB3\DATA\stocks1.csv')
stocks2=pd.read_csv('D:\Python nâng cao\Thu_Huyen_23174600074_17A2\DHKL17A2\LAB3\DATA\stocks2.csv')
stocks=pd.concat([stocks1,stocks2],ignore_index=True)
stocks
```

tạo MultiIndex

```python
print(stocks.index)
```

sdung groupby

```python
#tính trung bình mỗi ngày
daily_avg=stocks.groupby('date')[['open','high','low','close','volume']].mean().reset_index()
print("Trung bình các cột mỗi ngày là:")
print(daily_avg)
#tính trung bình mỗi mã 
code_avg=stocks.groupby('symbol')[['open','high','low','close','volume']].mean().reset_index()
print("giá trị trung bình mỗi mã chứng khoán là:")
print(code_avg)
```

sắp xếp

```python
stocks.sort_values(by=['date','symbol'],ascending=[True,True],inplace=True)
print(stocks)
```
