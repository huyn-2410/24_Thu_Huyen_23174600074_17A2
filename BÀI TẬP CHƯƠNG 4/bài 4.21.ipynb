```python
import pandas as pd
```

đọc dữ liệu

```python
drinks=pd.read_csv("drinks.csv",index_col=0)
drinks
```

phần 1:

```python
print("Kiểu dữ liệu ,kích thước và các cột là:")
print("Type:",type(drinks))
print("Shape:",drinks.shape)
print("Columns:",drinks.columns)
print("5 dòng đầu của dữ liệu là:")
drinks.head()
```

```python

print("\n5 dòng cuối của dữ liệu là:")
drinks.tail()
```

phần 2:

```python
print("Số bia trung bình theo từng châu lục:")
beer_mean=drinks.groupby('continent')['beer_servings'].mean()
beer_mean
```

phần 3:

```python
wine_sum=drinks.groupby('continent')['wine_servings'].sum()
print("Tổng số rượu vang tiêu thụ là:")
wine_sum
```

phần 5:

```python
stats=drinks.groupby('continent')[['beer_servings','wine_servings']].mean()
print("bia và rượu trung bình lần lượt là:")
stats
```

phần 5:

```python
median_values=drinks.groupby('continent')[['beer_servings','wine_servings']].median()
print("Trung vị cho cột bia và rượu là:")
median_values
```

phần 6:

```python
spirit=drinks.groupby('continent')['spirit_servings'].agg(['mean','max','min'])
print("số lượng rượu mạnh tiêu thụ trung bình ,lớn nhất,nhỏ nhất lần lượt là:")
spirit
```

phần 7:

```python
sorted_beer=drinks.sort_values(by='beer_servings',ascending=True)
print("Dữ liệu sắp xếp theo tổng lượng bia tiêu thụ giảm dần là:")
sorted_beer
```

5 quốc gia tiêu thu bia nhiều nhất và ít nhất

```python
print("5 quốc gia có lượng tiêu thụ bia ít nhất là:")
sorted_beer.head(5)
```

```python
print("5 quốc gia có số lượng tiêu thụ bia nhiều nhất là:")
sorted_beer.tail(5)
```
